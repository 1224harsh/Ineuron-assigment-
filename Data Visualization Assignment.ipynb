```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
irisdata=datasets.load_iris()
```

```python
irisdata.data
```

```python
print(irisdata.feature_names)
print(irisdata.target)
print(irisdata.target_names)
```

```python
x=decomposition.PCA(n_components=3).fit_transform(irisdata.data)
y=irisdata.target
```

```python
fig1=plt.figure(1,figsize=(10,8))
axes=Axes3D(fig1,elev=-150,azim=110)

axes.scatter(x[:,0],x[:,1],x[:,2],c=y,cmap=plt.cm.winter_r)

axes.set_title('Three Directions')
axes.set_xlabel('First Eigenvector')
axes.set_ylabel('Second Eigenvector')
axes.set_zlabel('Third Eigenvector')
```
